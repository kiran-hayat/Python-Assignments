# __name__

```python
__name__
```

```python
#                   __name__
"""`__name__ ak special variable ha python men , 
ye khud ba khud apny ap ko "__main__" ki value dyta ha , 
agr script ko direct run karna hu to , 
agr isy file men as a module import kia jay to ye 
 module ka name return karta ha """
```

```python
#                   __name__
"""`__name__` attribute module ko name dyti ha .
 jo file ka name ha  wo asal men __name__ attribute ki value ha.
isi trhan ye ap k  custom module ko name dyti ha. 
jesy k python module ko python ka name dyti ha """
```

```python
#              Purpose of __name__
"""the purpose of __name__ is to act as an import guard,
dosary alfaz men , jab ap koi python script banaty han to 
usy ap dosari file me import kar sagty han
bager zarort k ap usy program k tor py execute kar sagty han
"""

"""if __name__ =='__main__' idiom is ki zarort sirf tab parhti ha jab
1. ap apni file ko direct run karna chahty han , ya 
2. ap apni file ko module k tor py import karna chahty han """
```

### Example 

```python
if __name__ =='__main__':
    print('ye file direct run hu rahi ha.')
```

```python
import plotly
plotly.__name__
# yahan py plotly ki file as a module import kar k run ha rhai ha is liy ye file ka apna name return kar rahi ha
```

```python
# example 
import seaborn
seaborn.__dict__

# ye seaborn module k tamam objects ki attributes ko ak dictionary 
# men show kar kata ha , is se hamyn seaborn ya kisi bi module k
#  tamam objects ki attributes ak dictionary men show hu jaty han
```
